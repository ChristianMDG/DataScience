### Filtrez les données du data frame

```python
import pandas as pd
```

```python
data = pd.read_csv('clients.csv')
```

```python
data.head()
```

## Appliquez la sélection via des indices

### .iloc


```python
data.iloc[0,2]
```

```python
# sélectionner le nom des 10 premiers clients :
data.iloc[:10, 2]
```

```python
# sélectionner le genre des 10 premiers clients :
data.iloc[:10, 3]
```

```python
# sélectionner toutes les colonnes des 10 premiers clients :
data.iloc[:5, :]
```

```python
# sélectionner l’email et le nom des 10 clients suivants :
data.iloc[10:20, 1:3]
```

```python
# sélectionner l’e-mail et le genre des 10 derniers clients :
data.iloc[-10:, [1, 3]]
```

## Utilisez la sélection via des conditions

### .loc


```python
data.sort_values('nom').head()
```

```python
display(data.loc[0:10, :])
```

```python
display(data.iloc[0:10, :])
```

```python
# création d’un data frame trié pour ‘désordonner’ les Index
df_temp = data.sort_values('nom')
display(df_temp.head())
```

```python
# reset_index :
display(df_temp.reset_index().head())
```

```python
# reset_index sans garder les anciens Index :
display(df_temp.reset_index(drop=True).head())
```
