```python
import pandas as pd
```

```python
commandes = pd.read_csv('commande.csv')
commandes.head()
```

```python
produits = pd.read_csv('produits.csv')
produits.head()
```

```python
data = pd.merge(commandes,produits, on='id', how='left')
data.head()
```

```python
produit_plus_vendu = commandes.groupby('id')['nombre'].sum().idxmax()
produit_plus_vendu 
```

```python
data['CA'] = data['nombre'] * data['prix']
data['CA'].min()
```
